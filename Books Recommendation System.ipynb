# Popularity Based Recommendation System

```python
# merging 2 tables
rating_with_name = ratings.merge(books, on='ISBN')
```

```python
rating_with_name.head()
```

```python
# checking the number of ratings in the dataframe by grouping as per book title
num_rating_df = rating_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns = {'Book-Rating' : 'Num_ratings'}, inplace=True)
num_rating_df
```

```python
# average rating of the book
avg_rating_df = rating_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns = {'Book-Rating' : 'Avg_ratings'}, inplace=True)
avg_rating_df
```

```python
# creating one more variable
popular_df = num_rating_df.merge(avg_rating_df, on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['Num_ratings']>=250].sort_values('Avg_ratings', ascending=False).head(50)
```

```python
# top 50 books as per recommendation engine based on average ratings
popular_df
```

```python
books.columns
```

```python
# adding more details to the recommended books
popular_df = popular_df.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title',
                                                                                    'Book-Author',
                                                                                     'Year-Of-Publication', 'Publisher',
                                                                                    'Avg_ratings',
                                                                                    'Num_ratings','Image-URL-L']]
```

```python
# top 50 books as per recommendation engine based on average ratings with more columns
popular_df
```

# Collabarative Filtering Based Recommendation System


```python
# getting the headers
rating_with_name.head(1)
```

```python
rating_with_name.shape
```

```python
# creating a variable x, user based ratings
x = rating_with_name.groupby('User-ID').count()['Book-Rating'] > 200
educated_users = x[x].index
```

```python
educated_users
```

```python
filtered_rating = rating_with_name[rating_with_name['User-ID'].isin(educated_users)]
```

```python
filtered_rating.shape
```

```python
filtered_rating.head()
```

```python
# creating one more variable y, books based ratings
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
# top 50 books as per recommendation engine based on user ratings, with more than 50 book ratings
final_ratings
```

```python
# using pivot table
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt
```

```python
# replacing null values with zero
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#chances of user rating a book 
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores
```

```python
similarity_scores.shape
```

```python
books.columns
```

```python
# creating recommendation system
def recommendation_system(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    # using the similarity scores
    similar_items = sorted(list(enumerate(similarity_scores[index])), key = lambda x : x[1], reverse=True)[1:6]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        #pre processing
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-L'].values))
        
        data.append(item)
        
    return data
```

```python
# checking the recommendation system
recommendation_system('Harry Potter and the Prisoner of Azkaban (Book 3)')
```

```python
recommendation_system('Life of Pi')
```

```python
recommendation_system('Year of Wonders')
```
